# Introduction to Annual Income Analysis

Annual income analysis is a crucial aspect of understanding financial trends and patterns within a dataset. It involves examining the earnings of individuals or entities over a specific time period, usually a year. This analysis provides valuable insights into income distribution, disparities, and factors influencing financial well-being.

## Why Analyze Annual Income?

1. **Economic Insights**: Understanding income distribution within a population helps economists and policymakers make informed decisions regarding economic policies, taxation, and social welfare programs.

2. **Market Segmentation**: For businesses, analyzing annual income aids in identifying target markets, tailoring products or services, and setting pricing strategies.

3. **Financial Planning**: Individuals use income analysis to assess their own financial health, plan for the future, and make informed investment decisions.

4. **Socioeconomic Studies**: Researchers use income data to study socioeconomic trends, such as wealth gaps, mobility, and economic inequality.

## Analysis of Age and Income


In the following visual representation, a stacked bar chart illustrates the distribution of individuals based on their age groups and corresponding annual income. This analysis aims to discern the proportion of people earning more than 50k within each age bracket.

```python
bins = [18, 22, 27, 32, 37, 42, 47, 52, 57, 63, 69, 80, 100]
labels = ['18-22', '23-27', '28-32', '33-37', '38-42', '43-47', '48-52', '53-57', '58-63', '64-69', '70-80', '80+']

df['AgeRange'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

grouped_data = df.groupby(['AgeRange', 'income']).size().unstack(fill_value=0)

import matplotlib.pyplot as plt

grouped_data.plot(kind='bar', stacked=True)

plt.xlabel('Age Range')
plt.ylabel('Count')
plt.title('Distribution of Annual Income by Age Range')

plt.show()



```

## Native Country and Income
This bar chart categorizes individuals by their native country, providing a visual representation of the distribution of annual income levels within each country category. This analysis offers insights into income disparities across different regions.

```python
import pandas as pd
import matplotlib.pyplot as plt


region_mapping = {
    'North America': ['United-States', 'Canada'],
    'Europe': ['Germany', 'United-Kingdom', 'France', 'Italy', 'Spain'],
    'Asia': ['India', 'China', 'Japan', 'South', 'Philippines','Thailand'],
    'Central/South America': ['Mexico', 'Columbia', 'Ecuador', 'Peru', 'Puerto Rico'],
    'Other': ['Outlying-US(Guam-USVI-etc)', 'Haiti', 'Hungary', 'Holand-Netherlands']
}

df['Region'] = df['native-country'].apply(
    lambda country: next((region for region, countries in region_mapping.items() if country in countries), 'Other')
)

contingency_table = pd.crosstab(df['Region'], df['income'])

ax = contingency_table.plot(kind='bar', figsize=(10, 6), colormap='Set2')

plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Income Distribution by Region')

plt.show()
```

The bar charts below comapre the earnings of those from North America, and those from Asia

```python
specific_region = 'North America'

region_data = df[df['Region'] == specific_region]

contingency_table_specific = pd.crosstab(region_data['income'], region_data['Region'])

contingency_table_specific = contingency_table_specific.T

ax = contingency_table_specific.plot(kind='barh', figsize=(10, 6), colormap='Set2')

plt.xlabel('Count')
plt.ylabel('Income')
plt.title(f'Income Distribution for {specific_region}ns')

plt.show()

```

```python
specific_region = 'Asia'

region_data = df[df['Region'] == specific_region]

contingency_table_specific = pd.crosstab(region_data['income'], region_data['Region'])

contingency_table_specific = contingency_table_specific.T

ax = contingency_table_specific.plot(kind='barh', figsize=(10, 6), colormap='Set2')

plt.xlabel('Count')
plt.ylabel('Income')
plt.title(f'Income Distribution for {specific_region}ns')

plt.show()
```

It seems that Asians have a higher potential to earn more. This is perhaps due to the fact that Asians who immigrate to the USA tend to be skilled and higher educated.

```python
asian_data = df[df['Region'] == 'Asia']

asian_men = asian_data[asian_data['gender'] == 'Male']
asian_women = asian_data[asian_data['gender'] == 'Female']

contingency_table_men = pd.crosstab(asian_men['income'], asian_men['gender'])
contingency_table_women = pd.crosstab(asian_women['income'], asian_women['gender'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.bar(contingency_table_men.index, contingency_table_men['Male'], color='blue')
ax1.set_xlabel('Income')
ax1.set_ylabel('Count')
ax1.set_title('Income Distribution for Asian Men')

ax2.bar(contingency_table_women.index, contingency_table_women['Female'], color='purple')
ax2.set_xlabel('Income')
ax2.set_ylabel('Count')
ax2.set_title('Income Distribution for Asian Women')

plt.tight_layout()
plt.show()
```

```python
prof_school_data = df[df['education'] == 'Prof-school']

rich_prof_school_data = prof_school_data[prof_school_data['income'] == '>50K']

contingency_table_relationship = pd.crosstab(rich_prof_school_data['relationship'], rich_prof_school_data['income'])

total_husbands = contingency_table_relationship.loc['Husband'].sum()
total_wives = contingency_table_relationship.loc['Wife'].sum()

proportion_husbands = total_husbands / (total_husbands + total_wives)
proportion_wives = total_wives / (total_husbands + total_wives)

labels = ['Husbands', 'Wives']

plt.figure(figsize=(8, 8))
plt.pie([proportion_husbands, proportion_wives], labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'purple'])
plt.title('Proportion of Prof-School Graduates Earning >50K (Husbands vs. Wives)')

plt.axis('equal')  
plt.show()


```
